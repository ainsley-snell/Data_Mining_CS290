```python
import math
```

```python
# train so probailities equal 1, to reduce cross entropy loss (we wnat it to be 0#)
```

```python
#exponential functio
math.exp()
```

```python
x1=[1.9,-2.1,-0.3,0.4,-0.8]
x2=[0.3,2.5,-0.5,0.0,-1.1]
x3=[-0.6,0.2,1.4,-0.9,-0.2]
```

```python
def softmax(v):
  p=[]
  for i in  range(len(v)):
    p.append(math.exp( v[i]))
  total= sum(p)
  for i in range(len(v)):
    p[i]=p[i]/total
  return p
```

```python
probs1=softmax(x1)
print(probs1)
```

```python
probs2=softmax(x2)
print(probs2)
```

```python
probs3=softmax(x3)
print(probs3)
```
